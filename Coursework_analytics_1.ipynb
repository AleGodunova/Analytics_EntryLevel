Для того, чтобы проверить, что вы правильно соединили все датасеты в основной (без `students.csv`), запустите аналогичный приведенному ниже запрос. Выдача должна совпадать. Имейте в виду, что названия некоторых столбцов были переименованы.

```python
#Импортируем необходимые библиотеки
import numpy as np
import pandas as pd
import math
from datetime import datetime, timedelta
from dateutil.relativedelta import relativedelta
from pandas.tseries.offsets import MonthEnd, MonthBegin, Day

import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#Создаем основной датафрейм main
courses = pd.read_csv('Data\courses.csv', index_col=0)
contents = pd.read_csv('Data\course_contents.csv')
progress_phases = pd.read_csv('Data\progress_phases.csv')
progresses = pd.read_csv('Data\progresses.csv')


courses.rename(columns={'id':'course_id',\
                        'field':'course_field',\
                       'title':'course_title'}, inplace=True)
# Сортируем таблицу с курсами по разделам для удобства визуализации
courses = courses.sort_values('course_field').reset_index()

progresses.rename(columns={'id':'progress_id'}, inplace=True)


courses_df = pd.merge(courses,contents, on=('course_id'),  how='left')
progress_df = pd.merge(progress_phases, progresses, on=('progress_id'),  how='left')
main = pd.merge(progress_df, courses_df, on=['module_number', 'lesson_number', 'course_id'],  how='left')
```

```python
#Создаем основной датафрейм с датой рождения студентов
students_df = pd.read_csv('Data\students.csv')
students_df['birthday'] = pd.to_datetime(students_df['birthday'], dayfirst=False, errors='coerce')

# Удаляем пустые значения и выполняем преобразования для удобства
# использования датасета, а также для дальнейших вычислений
students_df.dropna(subset=['birthday'], inplace=True)
students_df.reset_index(drop=True, inplace=True)
students_df.drop(['id_'], axis='columns', inplace=True)
students_df.rename(columns={'id':'student_id'}, inplace=True)
students_df.head()
```

## Раздел 1

### Считаем количество курсов

```python
# Находим кол-во курсов путем применения метода .nunique(),
# который подсчитает кол-во уникальных значений в столбце
main.course_id.nunique()
```

### Количество модулей на каждом курсе

```python
# Группируем основной датафрейм по названию курсов,
# выбираем столбец с номерами модулей, считаем кол-во
# уникальных значений и выводим результат в виде датафрейма
main.groupby('course_title')['module_number'].nunique().to_frame()
```

### Количество уроков в каждом модуле на каждом курсе

```python
# Группируем основной датафрейм по названию курсов и модулей,
# выбираем столбец с номерами уроков,считаем кол-во
# уникальных значений и выводим результат в виде датафрейма
main.groupby(['course_title', 'module_title'])['lesson_number'].nunique().to_frame()
```

### Медианное количество уроков в модуле на каждом курсе

```python
# Группируем основной датафрейм по названию курсов и модулей,
# выбираем столбец с номерами уроков, считаем кол-во
# уникальных значений, для удобства осуществляем группировку
# по названию курсов, считаем медианное кол-во уроков 
# и выводим результат в виде датафрейма
main.groupby(['course_title', 'module_title'])['lesson_number']\
            .nunique().groupby('course_title').median().to_frame()
```

### Количество учеников на каждом курсе

```python
# Группируем основной датафрейм по названию курсов,
# выбираем столбец со студентами, считаем кол-во
# уникальных значений и выводим результат в виде датафрейма
main.groupby('course_title')['student_id'].nunique().to_frame()
```

### Минимальный, максимальный, средний, медианный возраст студентов

```python
# Вычисляем возраст студентов
today = pd.Timestamp.today()
age = lambda x: relativedelta(today, x).years
students_df['age'] = students_df.birthday.apply(age)

# Отсбрасываем выплески
students_df.drop(np.where(students_df.age.lt(10) | students_df.age.gt(80))[0],\
                 inplace=True)

# Выводим результат
students_df.age.apply(['min', 'max', 'mean', 'median']).to_frame()
```

### Минимальный, максимальный, средний, медианный возраст студентов на каждом курсе

```python
# Соединяем основной датафрейм и датафрейм со студентами
merged_main = main.merge(students_df, how='left', on='student_id')

# Выбираем необходимые для работы столбцы
# Группируем, а затем аггрегируем по требуемым значениям
merged_main[['course_title', 'age']].groupby('course_title')\
            .agg(['min', 'max', 'mean', 'median'])
```

### Bar-chart, отражающий количество студентов на каждом курсе.

```python
# Готовим новый датафрем для построения диаграммы
students_courses = merged_main.groupby('course_title')\
    .student_id.nunique().sort_values(ascending=False)

# Строим график
plt.figure(figsize=(10, 6))
plt.xticks(rotation=85)
plt.title('Количество студентов на каждом курсе')
plt.ylabel('Количество студентов, чел.')
plt.bar(students_courses.index, students_courses.values,\
       alpha=0.5)
plt.box(False)
plt.show()
```

### Горизонтальный (столбцы должны располагаться горизонтально) bar-chart, отражающий количество студентов на каждом курсе.

```python
# Создаем новый датафрейм, группируем по наменованиям и сфрере курсов
students_on_courses = merged_main.groupby(['course_title', 'course_field']).student_id\
        .nunique().sort_values(ascending=False).to_frame().reset_index(1)

# Рассчитываем медианное количество студентов по всем курсам
students_median = students_on_courses.student_id.median()

#Создаем диаграмму
fields_color = dict(zip(merged_main.course_field.unique(), plt.get_cmap('tab10',10).colors[[2,3,0,1]]))
plt.figure(figsize=(10,6))
ax = plt.axes()
plt.title('Количество студентов на каждом курсе')
plt.xticks(np.arange(0, max(students_courses+200), 200))
plt.tick_params(labelsize=10)
sns.barplot('student_id', students_on_courses.index,
            hue='course_field',
            hue_order=fields_color.keys(),
            dodge=False,
            data = students_on_courses, alpha=0.5,
            palette=fields_color)\
            .set(ylabel=None, xlabel=None)

# Добавляем медиану
plt.axvline(x=students_median, ls='-', label='Median', color='r') 
plt.legend(fontsize=15)
plt.show()
```

### Рассчитать минимальное, максимальное, среднее, медианное время прохождения каждого модуля для каждого курса.




```python
# Создаем новый датафрейм с записями по законченным домашним заданиям
module_duration = main[['course_id', 
                       'course_title', 
                       'module_number', 
                       'module_title', 
                       'start_date', 
                       'finish_date', 
                       'progress_id']]\
                        [(main['is_homework'] == True) & (main['status'] == 'done')]


# Форматируем колонки с датами
module_duration['finish_date'] = pd.to_datetime(module_duration['finish_date'])
module_duration['start_date'] = pd.to_datetime(module_duration['start_date'])

# Берем начало самой ранней и окончание самой поздней
# домашней работы
module_duration = module_duration.groupby(['course_id', 
                       'course_title', 
                       'module_number', 
                       'module_title',
                       'progress_id'])[['start_date', 'finish_date']]\
                        .agg({'start_date': 'min', 'finish_date': 'max'})

# Считаем время прохождения модуля в днях
module_duration['duration'] = (module_duration['finish_date']\
                               - module_duration['start_date'])\
                                .apply(lambda x: x.days)

# Некорректные значения удаляем и сбрасываем индекс
module_duration.drop(module_duration[module_duration['duration'] < 0].index, inplace=True)
module_duration.reset_index(inplace=True)

# Группируем по необходимым колонкам и аггрегруем
# по требуемым в условии значениям
duration = module_duration.groupby(['course_id', 
                        'course_title', 
                        'module_number', 
                        'module_title'])['duration'].agg(['min', 'max', 'mean', 'median'])

duration
```

### Расчет конверсии перехода студентов из одного модуля в другой на каждом курсе.

```python
# Создаем новый датафрейм с конверсией
# и выбираем значения из колонки is_homework
# с признаком True
course_hw = main[['course_id',\
                'module_number',\
                'lesson_number']]\
                [(main['is_homework'] == True)]\
                .sort_values(by='lesson_number')

# Находим первый и последний модуль в курсах
course_last_hws = course_hw.groupby(['course_id',\
                'module_number'], as_index=False).last()
course_first_hws = course_hw.groupby(['course_id',\
                'module_number'], as_index=False).first()

# Сброшены индексы и упорядочены данные,
# удалены первые модули, т.к. они не удовлетворяют условию
course_first_hws.index = course_first_hws.index - 1
course_cur_prev_hw = course_last_hws.merge(course_first_hws,\
                left_index=True,\
                right_index=True,\
                suffixes=('_prev', '_cur'))
course_cur_prev_hw = course_cur_prev_hw[course_cur_prev_hw\
                ['course_id_prev'] == course_cur_prev_hw['course_id_cur']]
course_cur_prev_hw.drop('course_id_cur', axis='columns', inplace=True)
course_cur_prev_hw.rename(columns={'course_id_prev': 'course_id'}, inplace=True)

# Выбираем законченные ДЗ в разрезе уроков, группируем,
# считаем количество прогрессов
df_count_finished_lessons = main[(main['is_homework'] == True) & (main['status'] == 'done')]\
                                .groupby(['course_id', 
                                          'module_number', 
                                          'module_title', 
                                          'lesson_number'])\
                                ['progress_id'].count().reset_index()

# Выбираем все ДЗ (начатые, законченные, в ожидании)
# в разрезе уроков, группируем, считаем количество прогрессов
df_count_started_lessons = main[(main['is_homework'] == True)]\
                                .groupby(['course_id',
                                          'module_number', 
                                          'module_title', 
                                          'lesson_number'])\
                                ['progress_id'].count().reset_index()

# К собранной таблице с нужными номерами модулей и уроков
# присоединяем данные по количеству законченных ДЗ
modules_conversion_df = course_cur_prev_hw.merge(df_count_finished_lessons, 
                         how='left', 
                         left_on=['course_id', 'module_number_prev', 'lesson_number_prev'], 
                         right_on=['course_id', 'module_number', 'lesson_number'])

# К собранной таблице с нужными номерами модулей и уроков присоединяем данные по количеству начатых ДЗ (возможно законченных)
modules_conversion_df = modules_conversion_df.merge(df_count_started_lessons, 
                         how='left', 
                         left_on=['course_id', 'module_number_cur', 'lesson_number_cur'], 
                         right_on=['course_id','module_number', 'lesson_number'])

# Считаем конверсию
modules_conversion_df['conversion'] = modules_conversion_df['progress_id_y']\
                        / modules_conversion_df['progress_id_x']

# оставляем только нужные колонки и переименовываем их
conversion_df = modules_conversion_df[['course_id',\
                                       'module_number_cur',\
                                       'module_title_y',\
                                       'conversion']]
conversion_df.columns = ['course_id', 'module_number',\
                         'module_title', 'conversion']
conversion_df.fillna(value={'conversion': 0}, inplace=True)
conversion_df
```

Иногда студенты берут курсы, которые оказываются для них неподъемными. Это может быть как по причинам недостаточной изначальной подготовки, так и по причинам, связанным с низкой мотивацией студента. Для улучшения качества контента полезно выявить причину. На основании имеющихся данных придумайте метрику успеваемости студента. **Обоснуйте ее.** Выявите таких студентов на каждом курсе, чтобы предоставить продюсерам список проблемных студентов. 

```python
#Создаем новый датафрейм
performance_df = main.copy()

# Добавляем новый столбец, куда добавим
# законченные уроки
performance_df['complete'] = performance_df['status'] == 'done'
performance_df['complete'] = performance_df['complete'].astype(int)

# Группируем дф по каждому студенту и аггрегируем
# по количеству статусов уроков и сумме завершенных
# уроков
performance_df = performance_df.groupby('student_id')\
                .agg({'status':'count', 'complete':'sum'})

# Создаем столбец с успеваемостью, методом отношения
# Количества завершенных уроков к начатым
performance_df['performance'] = performance_df['complete']\
                /performance_df['status']

# Выполняем окончательную сортировку
performance_df.sort_values(by=['performance','status'],\
                ascending=[True, False]).reset_index()
```
